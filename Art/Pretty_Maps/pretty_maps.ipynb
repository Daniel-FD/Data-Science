```python
import prettymaps
import pprint
import geopandas as gpd
from shapely.geometry import Polygon
import osmnx as ox
```

**your_query** can be:
1. An address (Example: "Porto Alegre"),
2. Latitude / Longitude coordinates (Example: (-30.0324999, -51.2303767))
3. A custom boundary in GeoDataFrame format

```python
lon_lat_list = [[ -8.433745660425185, 43.347331541948869 ], 
                [ -8.372057143421673, 43.347009482909961 ], 
                [ -8.371592516397957, 43.392029306070974 ], 
                [ -8.433326671648997, 43.392351869045932 ], 
                [ -8.433745660425185, 43.347331541948869 ]]
```

```python
polygon_geom = Polygon(lon_lat_list)
polygon = gpd.GeoDataFrame(index=[0], crs='epsg:4326', geometry=[polygon_geom]) 
polygon.to_file(filename='polygon_input.geojson', driver='GeoJSON')
```
