```python
from optimize_prompt import optimize_prompt
import pandas as pd
```

```python
# Load the evaluation dataset
df = pd.read_csv('reviews.csv', encoding='ISO-8859-1')

# Preprocess the dataset
eval_data = df[['Text', 'Sentiment']]
eval_data = eval_data.rename(columns={'Text': 'text', 'Sentiment': 'label'})

# Randomly select 50 positive and 50 negative samples
positive_samples = eval_data[eval_data['label'] == 1].sample(n=5, random_state=42)
negative_samples = eval_data[eval_data['label'] == 0].sample(n=5, random_state=42)

# Combine the samples and shuffle
eval_data = pd.concat([positive_samples, negative_samples]).sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
initial_prompt = (
        "You are a sentiment analysis classifier. Determine whether the provided text expresses a positive sentiment about a moview review "
        "Respond with '1' if it is positive, or '0' if it is negative. Do NOT answer anything else."
    )
# Number of optimization iterations
iterations = 3  # You can adjust this value
# Start the prompt optimization process
optimize_prompt(initial_prompt, eval_data, iterations)
```
